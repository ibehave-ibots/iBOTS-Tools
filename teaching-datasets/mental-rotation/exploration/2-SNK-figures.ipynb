```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
data_path = '../data/MentalRotation.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
# Time taken with respect to age 

fig = plt.figure(figsize=(25,10))
sns.lineplot(data=df,x='Age',y='Time',hue='Correct')
plt.ylabel('Time [ms]',fontsize=15)
plt.xlabel('Age',fontsize=15)
```

```python
len(df[df['Correct']==1])
```

```python
# separated to 1 and 0 for visibility (imbalanced data)
sns.kdeplot(data=df[df['Correct']==1],x='Time')
sns.kdeplot(data=df[df['Correct']==0],x='Time')
```

```python
# separated to 1 and 0 for visibility (imbalanced data)
sns.kdeplot(data=df[df['Correct']==1],x='Age')
sns.kdeplot(data=df[df['Correct']==0],x='Age')
```
